```python
import scipy.stats as sts
from scipy.stats import shapiro

for i in data.columns:
  stats,p_value=shapiro(data[i])
  alpha=0.05
  #h0: Samples is come from normally distributed
  #ha: Sample doesn't come from normally distributed
  if p_value > alpha:
    print('Normally Distributed ,Fail to reject ho',i)
    data[i].fillna(data[i].mean(),inplace=True)
  else:
    print('Skeewed data, Reject the ho',i)
    data[i].fillna(data[i].median(),inplace=True)
```

```python
#splitting the dataset into train and test

X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
y_test
```

```python
#standarlizing the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_test
```

#Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
#print the coeff
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
#On which parameter model has been train
print(regression.get_params())
```

```python
#predict on test data

pred=regression.predict(x_test)
```

#Assumptions

```python
# visualizing the test and predict data
sns.scatterplot(x=y_test,y=pred)
```

```python
#Residulas
residuals=y_test-pred
print(residuals)
```

```python
sns.displot(residuals,kind='kde')
```

```python
#scatter plot with respect to predict and residuals
sns.scatterplot(y=residuals,x=pred)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_squared_error(y_test,pred))
print(mean_absolute_error(y_test,pred))
```

#R Square And Adjusted R square

Formula

  R^2=1-SSR/SST  

  R^2=coeffecient od determination

  SSR= sum of square of residual
  
  SST=total sum of square

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,pred)
print(score)
```

#Adjusted R2 = 1  [(1-R2)*(n-1)/(n-k-1)]

where:

R2: The R2 of the model n: The number of observations k: The number of predictor variables

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
